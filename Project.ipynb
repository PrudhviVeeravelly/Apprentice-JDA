```python
import numpy as np,pandas as pd
d=pd.read_csv("C:/Users/admin/Desktop/Apprentice-JDA/clickstreampath.csv")
df=pd.DataFrame(d)
dff=df.values.tolist()
dff
```

```python
dff2=[]
for line in dff:
     sample2=[]
     for ind in line:
        ind = str(ind)
        if ('nan' in ind) or ('user' in ind):
            pass
        else:
            sample2.append(ind)
        dff2.append(sample2)
```

```python
dff2
```

```python
from markovclick.models import MarkovClickstream
m = MarkovClickstream(dff2)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(17,8))
sns.heatmap(m.prob_matrix, xticklabels=m.pages, yticklabels=m.pages)
```
